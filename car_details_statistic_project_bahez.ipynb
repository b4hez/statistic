# Car Details Dataset Analysis

This notebook analyzes a dataset of used cars with the following features:

- Car name
- Year of manufacture
- Kilometers driven
- Selling price
- Fuel type

Weâ€™ll clean the data, explore key trends, and visualize relationships to better understand factors affecting car prices.

```python
import matplotlib.pyplot as plt
%matplotlib inline

mydata.hist(figsize=(30,40))
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x="year", y="selling_price", data=mydata)
```

```python
pd.crosstab(mydata['year'],mydata['km_driven'] )
```

```python
pd.crosstab(mydata['name'],mydata['km_driven'] )
```

```python
sns.countplot(x="km_driven", hue="year", data=mydata)
```

```python
pd.pivot_table(mydata, index=['name', 'seller_type'],
                     columns=[ 'fuel'], aggfunc=len)
```

```python
pd.pivot_table(mydata, values='year', index=['name'],
                     columns=[ 'fuel'], aggfunc='mean') # or any other applicable numeric column like 'km_driven'
```
